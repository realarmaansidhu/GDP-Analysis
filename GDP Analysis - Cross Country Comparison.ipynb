```python
import numpy as np
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
from plotly import express as px
from plotly import offline as pyo
```

```python
df = pd.read_csv("gdp.csv")
```

```python
df["GDP Growth Rate"] = round(df.groupby("Country Name")["Value"].pct_change() * 100, 2)
df["GDP Growth Rate"] = df["GDP Growth Rate"].replace(np.NaN, 0)
df
```

```python
fig_a = px.line(df, x = "Year", y = "Value", color = "Country Name")
pyo.plot(fig_a, filename = "GDP - All.html", auto_open = False)
```

```python
df_c = df[df["Country Name"] == "China"]
df_i = df[df["Country Name"] == "India"]
df_ci = pd.concat([df_c, df_i], axis = 0)
df_ci
```

```python
fig_ci = px.line(df_ci, x = "Year", y = "Value", color = "Country Name", title = "GDP Comparison | India & China")
pyo.plot(fig_ci, filename = "GDP Comparison - IND|CHN.html", auto_open = False)
```

```python
df_w = df[df["Country Name"] == "World"]
df_cw = pd.concat([df_c, df_w], axis = 0)
df_cw
```

```python
fig_cw = px.line(df_cw, x = "Year", y = "Value", color = "Country Name", title = "GDP Comparison | World & China")
pyo.plot(fig_cw, filename = "GDP Comparison - WLD|CHN.html", auto_open = False)
```
