## GDP Growth on whole Dataset

```python
df["GDP Growth Rate"] = round(df.groupby("Country Name")["Value"].pct_change() * 100, 2)
df["GDP Growth Rate"] = df["GDP Growth Rate"].replace(np.NaN, 0)
df
```

```python
df.groupby(by=["Country Name"])["GDP Growth Rate"].max()
```

```python
df.groupby(by=["Country Name"])["GDP Growth Rate"].mean()
```
