## Analyzing Arab World

```python
df_pr = df[df["Country Name"] == "Arab World"]
sns.set_theme()
df_pr.plot(kind="line", x = "Year", y = "Value", legend = True, xlabel = "YEARS", ylabel = "GDP")
plt.show()
```

```python
dc_pr = {}
for index, row in df_pr.iterrows():
    key = row["Year"]
    value = row["Value"]
    dc_pr[key] = value

dc_pr
```

```python
gr = {}
for year in df_pr["Year"]:
    if year == 1968:
        gr[year] = 0
    else:
        gr[year] = (((dc_pr[year] - dc_pr[year-1])/dc_pr[year-1])*100)
gr
```

```python
df_cr = pd.DataFrame(columns = ["Year", "Value"])
df_cr["Year"] = gr.keys()
df_cr["Value"] = gr.values()
df_cr["Value"] = round(df_cr["Value"], 2)
df_cr
```

```python
df_cr.plot(kind = "line", x = "Year", y = "Value", xlabel = "YEAR", ylabel = "GDP", legend = True)
plt.show()
```

```python
df_pr["GDP Growth Rate"] = df_cr["Value"]
df_pr
```
